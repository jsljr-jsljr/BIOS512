Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.res = 300, repr.plot.height = 7, repr.plot.width = 6.5)

q = ggplot(penguins, aes(x = body_mass_g))
q = q + geom_histogram(bins = 50, col = "black", fill='steelblue', alpha=0.75)
q = q + theme_bw()
q = q + facet_wrap(~year, ncol = 1)
q = q + scale_x_continuous(labels = comma)
q = q + labs(x ="Body mass in grams", y = "Number of penguins")
q
```

Is there a difference between the body mass of male and female penguis? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to see compare within each species-group.**

```r
r = ggplot(penguins, aes(x = body_mass_g, y = sex))
r = r + geom_point(position = position_jitter(h = 0.15 , w = 0))
r = r + theme_bw()
r = r + theme(panel.border = element_blank())
r = r + facet_wrap(~species, ncol=1)
r = r + labs(x = "Body mass in grams", y = "")
r
```
